```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATASET_DIR = '/Users/stepanmazokha/Downloads/GlobecomPOWDER'

FILE_BIN = os.path.join(DATASET_DIR, '4G_Day_1_bes_s1.bin')
FILE_JSON = os.path.join(DATASET_DIR, '4G_Day_1_bes_s1.json')
```

```python
def read_json(file_path):
    with open(file_path) as f:
        text = json.load(f)
    return text

def read_powder_file(path_signal, path_metadata):
    iq = np.fromfile(path_signal, dtype=np.complex64)
    metadata = read_json(path_metadata)

    return iq, metadata
```

```python
iq, metadata = read_powder_file(FILE_BIN, FILE_JSON)
```

```python
plt.figure(figsize=(20, 6), dpi=100)
plt.plot(np.real(iq))
plt.title(f'IQ Samples from 4G_Day_1_bes_s1.json')
plt.xlabel('Amplitude')
plt.ylabel('Sample IDX')
```
